```python
def recommand(movie):
    # Find the index/position of the input movie in our dataset
    index = new_data[new_data['title'] == movie].index[0]
    
    # Create a sorted list of (movie_index, similarity_score) tuples
    # sorted from most similar to least similar
    distance = sorted(
        enumerate(similarity[index]),  # Add indices to similarity scores
        key=lambda vector: vector[1],   # Sort by similarity score (2nd element)
        reverse=True                     # Highest similarity first
    )
    
    # Print top 5 similar movies (skip index 0 as it's the input movie itself)
    for i in distance[1:6]:
        print(new_data.iloc[i[0]].title)  # Get and print movie title
```

```python
# Test with various movies and manually check if recommendations make sense
test_movies = ['Iron Man', 'The Godfather', 'Titanic', 'Toy Story', 'The Dark Knight']

for movie in test_movies:
    print(f"\nRecommendations for: {movie}")
    print("-" * 50)
    recommand(movie)
```

```python
def evaluate_genre_overlap(movie_title, top_n=5):
    """
    Calculates what percentage of recommended movies share at least one genre 
    with the input movie
    """
    # Get the input movie's index
    index = new_data[new_data['title'] == movie_title].index[0]
    
    # Get input movie's genres from the original data
    input_genres = set(data.iloc[index]['genre'].split(','))
    
    # Get top N recommendations
    distance = sorted(enumerate(similarity[index]), key=lambda x: x[1], reverse=True)
    
    # Count how many recommendations share genres
    genre_matches = 0
    print(f"\nEvaluating recommendations for: {movie_title}")
    print(f"Input movie genres: {input_genres}\n")
    
    for i in distance[1:top_n+1]:
        rec_title = new_data.iloc[i[0]]['title']
        rec_genres = set(data.iloc[i[0]]['genre'].split(','))
        overlap = input_genres & rec_genres
        
        if overlap:
            genre_matches += 1
            print(f"✓ {rec_title} - Shared: {overlap}")
        else:
            print(f"✗ {rec_title} - No overlap")
    
    score = (genre_matches / top_n) * 100
    print(f"\nGenre Overlap Score: {score}%")
    return score

# Test it
evaluate_genre_overlap('Iron Man')
evaluate_genre_overlap('The Godfather')
evaluate_genre_overlap('Titanic')
```
